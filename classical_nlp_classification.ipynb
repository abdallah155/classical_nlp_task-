```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
nltk.download('punkt')
nltk.download('wordnet')
```

```python
def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    tokens = nltk.word_tokenize(text)
    lemmatized_text = [lemmatizer.lemmatize(token, wordnet.VERB) for token in tokens]
    return ' '.join(lemmatized_text)
```

```python
df['Article Title'] = df['Article Title'].apply(lemmatize_text)
df['Article Title'].values
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Article Title'])
bag_of_words = X.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(df['Article Title'])
tfidf_bag_of_words = X_tfidf.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(df['Article Title'])

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, df['Category'], test_size=0.2, random_state=42)

# Train Logistic Regression classifier
classifier = LogisticRegression()
classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = classifier.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
```

```python
import gensim.downloader as api

# Load pre-trained word embeddings
word_vectors = api.load("glove-wiki-gigaword-100")

def sentence_embedding(sentence):
    words = sentence.split()
    word_embeddings = [word_vectors[word] for word in words if word in word_vectors]
    if not word_embeddings:
        return np.zeros(word_vectors.vector_size)
    return np.mean(word_embeddings, axis=0)
```

```python
df['Glove'] = df['Article Title'].apply(sentence_embedding)
df['Glove']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(df['Glove'], df['Category'], test_size=0.2, random_state=42)

# Train Logistic Regression classifier
classifier = LogisticRegression()
classifier.fit(X_train.tolist(), y_train)

# Make predictions on the test set
y_pred = classifier.predict(X_test.tolist())

# Evaluate the model
print(classification_report(y_test, y_pred))
```

```python
model_name = "fasttext-wiki-news-subwords-300"
model = api.load(model_name)

# Encode text data using FastText model
def encode_text(text):
    tokens = text.split()
    vectors = [model[word] for word in tokens if word in model]
    if vectors:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)

df["fastt"] = df["Article Title"].apply(encode_text)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(np.stack(df["fastt"]), df["Category"], test_size=0.2, random_state=42)

# Train a classifier
classifier = LogisticRegression()
classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = classifier.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
```
